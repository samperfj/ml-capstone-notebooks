```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data
data = {
    'model': ['Baseline', 'KNN_item', 'KNN_user', 'NMF', 'SVD', 'NN_Embedding', 'Custom_NN_Embedding'],
    'RMSE': [1.910, 1.289, 1.283, 1.289, 1.289, 0.119, 0.094]
}
df = pd.DataFrame(data)

# Set custom colors: 
blue_color = '#1f77b4'  # A more visually appealing blue
green_color = '#2ca02c'  # A smoother green
orange_color = '#ff7f32'  
colors = [blue_color, blue_color, blue_color, orange_color, orange_color, green_color, green_color]

# Create bar plot
plt.figure(figsize=(10, 6))
bars = plt.bar(
    df['model'],
    df['RMSE'],    # Y-axis
    color=colors,  # Bar color
    width=0.4      # Adjusted width for better spacing
)

# Label formatting
plt.xlabel('')
plt.ylabel('RMSE', fontsize=12)
plt.title('Model RMSE comparison', fontsize=14)

# Add RMSE value on top of each bar
for bar, rmse in zip(bars, df['RMSE']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), 
             f'{rmse:.3f}', ha='center', va='bottom', fontsize=10)

# Reduce the font size of the x-axis labels
plt.xticks(df['model'], fontsize=8, ha='center')

# Increase the space between top of bars and plot top border
plt.ylim(0, max(df['RMSE']) + 0.15)  # Increase the y-axis upper limit for more space

# Tight layout to ensure everything fits
plt.tight_layout()

# Show plot
plt.show()
```
